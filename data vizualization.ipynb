```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
x = pd.read_csv('https://github.com/basseat/Datasets/raw/master/Standard_Metropolitan_Areas_Data-data.csv')
x.head()
```

```python
##scatter plot
plt.scatter(x.crime_rate, x.percent_senior)
plt.show()
```

```python
##line plot
plt.plot(x.work_force, x.income)
plt.show()
```

```python
##histogram
plt.hist(x.percent_senior)
plt.show()
```

```python
##vertical bar plot
plt.bar(x.region, x.crime_rate)
plt.show()
```

```python
##horizontal bar plot
plt.barh(x.region, x.crime_rate)
plt.show()
```

```python
##pyplot formatting
plt.plot(x.work_force, x.income, color = 'r', label = 'workforce')
plt.plot(x.physicians, x.income, label = 'physicians')
plt.legend()

plt.show()
```

```python
plt.scatter(x.percent_senior, x.crime_rate)
plt.title('plot of crime rate vs percent senior')
plt.xlabel('percent senior')
plt.ylabel('crime rate')
plt.show()
```

```python
##formatting the plot
plt.plot(x.physicians, x.income, color = 'blue',  linestyle = 'dashed', marker = 'o')
plt.plot(x.work_force, x.income, color = 'c', linestyle = 'dotted', marker = 'x')
plt.plot(x.hospital_beds, x.income, color = '0.75', linestyle = 'dashdot', marker = '^')
plt.show()
```

```python
##Subplots
plt.subplot(1,2,1)
plt.plot(x.work_force, x.income, 'go')
plt.title('income vs work force')

plt.subplot(1,2,2)
plt.plot(x.hospital_beds, x.income, 'r^')
plt.title('income vs hospital beds')

plt.suptitle('sub plots')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey = True)

fig.suptitle('horizontally stacked subplots')
ax1.plot(x.work_force, x.income, 'go')
ax1.set_title('income vs work force')

ax2.plot(x.hospital_beds, x.income, 'r^')
ax2.set_title('income vs hospital beds')

plt.show()

```

```python
fig, ax = plt.subplots(2,2, figsize = (6,6))
ax[0,0].plot(x.work_force, x.income, 'go')
ax[0,1].plot(x.work_force, x.income, 'bo')
ax[1,0].plot(x.work_force, x.income, 'yo')
ax[1,1].plot(x.work_force, x.income, 'ro')

plt.show()
```
